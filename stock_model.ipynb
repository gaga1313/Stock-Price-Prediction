```python
data=data.dropna()
```

```python
X=data.drop(["Close" , "Date"] , axis=1)
Y=pd.DataFrame(data=data.Close , columns=['Close'])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size=0.3 , random_state=42)
x_train.columns
```

```python
x_train=pd.DataFrame(data=x_train , columns=['Open', 'High', 'Low', 'Adj Close', 'Volume'])
x_test=pd.DataFrame(data=x_test , columns=['Open', 'High', 'Low', 'Adj Close', 'Volume'])
```

```python
x_train=x_train.drop(['Volume'  ]  , axis=1)
x_test=x_test.drop(['Volume'  ]  , axis=1)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

regr.fit (x_train, y_train)
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_hat)
```
